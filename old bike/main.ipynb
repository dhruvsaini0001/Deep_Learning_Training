```python
val_count_bike = df['brand'].value_counts()
condition_reduction = val_count_bike[val_count_bike>30]
df['brand'] = df['brand'].apply(lambda x : x if x in  condition_reduction else 'others' )
```

```python
from sklearn.preprocessing import OneHotEncoder,LabelEncoder
```

```python
label_encoder_model = LabelEncoder()
df['model_name'] = label_encoder_model.fit_transform(df['model_name'])
```

```python
df.head()
```

```python
onehot_encoder_brand=OneHotEncoder(handle_unknown='ignore')
brand_encoded = onehot_encoder_brand.fit_transform(df[['brand']]).toarray()
brand_encoded_df = pd.DataFrame(brand_encoded,columns=onehot_encoder_brand.get_feature_names_out(['brand']))
```

```python
brand_encoded_df 

```

```python
df = pd.concat([df.drop('brand',axis=1),brand_encoded_df],axis=1)
```
